```python
from matplotlib import pyplot as pl
import numpy as np
```

# Seno Cosseno de $-\pi$ a $\pi$, com 256 pontos

```python
# 256 pontos de -pi a pi
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
# cosseno e seno desses pontos
C,S = np.cos(X), np.sin(X)
```

```python
fig = pl.figure()
# adiciona um plot à figura de dimensões 1x1, ocupando a posição 1
ax = fig.add_subplot(111)
# plota cosseno
ax.plot(X, C)
# plota seno
ax.plot(X, S)
```

# Mudando o tamanho da figura

```python
# tamanho em inch
fig = pl.figure(figsize=(3,2))
ax = fig.add_subplot(111)
ax.plot(X, C)
ax.plot(X, S)
```

# Mudando características da curva

```python
fig = pl.figure()
ax = fig.add_subplot(111)
ax.plot(X, C, color='green', linewidth=2.0, linestyle='-')
ax.plot(X, S, color='blue', linewidth=1.0, linestyle=':')
```

### Similar

```python
fig = pl.figure()
ax = fig.add_subplot(111)
ax.plot(X, C, 'g-', linewidth=2.0)
ax.plot(X, S, 'b:')
```

# Setando os limites

```python
fig = pl.figure()
ax = fig.add_subplot(111)
ax.plot(X, C)
ax.plot(X, S)

# limites da coordenada x
ax.set_xlim(X.min()*1.1, X.max()*1.1)
# limites da coordenada y
ax.set_ylim(C.min()*1.1, C.max()*1.1)
```

# Setando os ticks

```python
fig = pl.figure()
ax = fig.add_subplot(111)
ax.plot(X, C)
ax.plot(X, S)
ax.set_xlim(-3.0, 3.0)
ax.set_ylim(-.5, 1.0)

# posição dos ticks para coordenadas x
ax.set_xticks(np.linspace(-3.0, 3.0, 10, endpoint=True))
# posição dos ticks para coordenadas y
ax.set_yticks(np.linspace(-.5, 1.0, 10, endpoint=True))
```

### Com labels

```python
fig = pl.figure()
ax = fig.add_subplot(111)
ax.plot(X, C)
ax.plot(X, S)
ax.set_xlim(-4.0, 4.0)
ax.set_ylim(-1.0, 1.0)

# posição dos ticks
xticks = [-np.pi, -np.pi/2, 0, np.pi/2, np.pi]
# label dos ticks
xticks_labels = ['$-\pi$', '$-\frac{\pi}{2}$', '0', '$\pi/2$', '$\pi$']

ax.set_xticks(xticks)
ax.set_xticklabels(xticks_labels)
ax.set_yticks(np.linspace(-1.0, 1.0, 10, endpoint=True))
```

# Salvando a figura

```python
fig = pl.figure()
ax = fig.add_subplot(111)
ax.plot(X, C)
ax.plot(X, S)

pl.savefig('filename.png')
```

# Movendo os spines

```python
fig = pl.figure()
ax = fig.add_subplot(111)
ax.plot(X, C)
ax.plot(X, S)

# deixa os spines do topo e da direita invisíveis
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# move os spines do bottom e da esquerda
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# remove os ticks no topo e à esquerda
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
```

# Adicionando legenda

```python
fig = pl.figure()
ax = fig.add_subplot(111)
# necessário nomear
ax.plot(X, C, label='cosseno')
ax.plot(X, S, label='seno')

# seta a posição
ax.legend(loc='upper left')
```

# Anotando pontos

```python
t = 2*np.pi/3

fig = pl.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.plot(X, C)
ax.plot(X, S)

# adiciona linhas conectando coordenada x ao ponto da curva do cosseno
ax.plot(np.ones(10)*t, np.linspace(cos(t), 0, 10), 'b--')
# anota
# xy : posição do ponto anotado
# xytext : posição da anotação
# arrowpropos : propriedades do arrow da anotação ao ponto
# ver: http://matplotlib.org/1.2.1/users/annotations_guide.html
ax.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$', 
            xy=(t, np.cos(t)),
            xytext=(t+.2, np.cos(t)-.1),
            fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
# adiciona linhas conectando coordenadas x ao ponto da curva do seno
ax.plot(np.ones(10)*t, np.linspace(0, sin(t), 10), 'g--')
# anota
ax.annotate(r'$\sin(\frac{2\pi}{3}=\frac{\sqrt{3}}{2}$',
            xy=(t, np.sin(t)),
            xytext=(t+.2, np.sin(t)+.2), 
            fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2"))

#             movendo spins
# deixa os spines do topo e da direita invisíveis
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# move os spines do bottom e da esquerda
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# remove os ticks no topo e à esquerda
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
```
