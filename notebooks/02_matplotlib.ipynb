```python
from matplotlib import pyplot as pl
import numpy as np
```

# Seno Cosseno de $-\pi$ a $\pi$, com 256 pontos

```python
# 256 pontos de -pi a pi
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
# cosseno e seno desses pontos
C,S = np.cos(X), np.sin(X)
```

```python
fig = pl.figure()
# adiciona um plot à figura de dimensões 1x1, ocupando a posição 1
ax = fig.add_subplot(111)
# plota cosseno
ax.plot(X, C)
# plota seno
ax.plot(X, S)
```

# Mudando o tamanho da figura

```python
# tamanho em inch
fig = pl.figure(figsize=(3,2))
ax = fig.add_subplot(111)
ax.plot(X, C)
ax.plot(X, S)
```

# Mudando características da curva

```python
fig = pl.figure()
ax = fig.add_subplot(111)
ax.plot(X, C, color='green', linewidth=2.0, linestyle='-')
ax.plot(X, S, color='blue', linewidth=1.0, linestyle=':')
```

# Setando os limites

```python
fig = pl.figure()
ax = fig.add_subplot(111)
ax.plot(X, C)
ax.plot(X, S)

# limites da coordenada x
ax.set_xlim(X.min()*1.1, X.max()*1.1)
# limites da coordenada y
ax.set_ylim(C.min()*1.1, C.max()*1.1)
```

# Setando os ticks

```python
fig = pl.figure()
ax = fig.add_subplot(111)
ax.plot(X, C)
ax.plot(X, S)
ax.set_xlim(-3.0, 3.0)
ax.set_ylim(-.5, 1.0)

# posição dos ticks para coordenadas x
ax.set_xticks(np.linspace(-3.0, 3.0, 10, endpoint=True))
# posição dos ticks para coordenadas y
ax.set_yticks(np.linspace(-.5, 1.0, 10, endpoint=True))
```

### Com labels

```python
fig = pl.figure()
ax = fig.add_subplot(111)
ax.plot(X, C)
ax.plot(X, S)
ax.set_xlim(-4.0, 4.0)
ax.set_ylim(-1.0, 1.0)

# posição dos ticks
xticks = [-np.pi, -np.pi/2, 0, np.pi/2, np.pi]
# label dos ticks
xticks_labels = ['$-\pi$', '$-\frac{\pi}{2}$', '0', '$\pi/2$', '$\pi$']

ax.set_xticks(xticks)
ax.set_xticklabels(xticks_labels)
ax.set_yticks(np.linspace(-1.0, 1.0, 10, endpoint=True))
```

# Movendo os spines

```python
fig = pl.figure()
ax = fig.add_subplot(111)
ax.plot(X, C)
ax.plot(X, S)

# deixa os spines do topo e da direita invisíveis
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# move os spines do bottom e da esquerda
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# remove os ticks no topo e à esquerda
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
```

# Anotando pontos

```python
t = 2*np.pi/3

fig = pl.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.plot(X, C, linewidth=2.5)
ax.plot(X, S, linewidth=2.5)

# adiciona linhas conectando coordenada x ao ponto da curva do cosseno
ax.plot(np.ones(10)*t, np.linspace(cos(t), 0, 10), 'b--')
# ponto
ax.scatter([t,], [np.cos(t),], 50)
# anota
# xy : posição do ponto anotado
# xytext : posição da anotação
# arrowpropos : propriedades do arrow da anotação ao ponto
# ver: http://matplotlib.org/1.2.1/users/annotations_guide.html
ax.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$', 
            xy=(t, np.cos(t)),
            xycoords='data',
            xytext=(t+.3, np.cos(t)-.2),
            fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
# adiciona linhas conectando coordenadas x ao ponto da curva do seno
ax.plot(np.ones(10)*t, np.linspace(0, sin(t), 10), 'g--')
ax.scatter([t,], [np.sin(t),], 50)
# anota
ax.annotate(r'$\sin(\frac{2\pi}{3}=\frac{\sqrt{3}}{2}$',
            xy=(t, np.sin(t)),
            xytext=(t+.2, np.sin(t)+.2), 
            fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2"))

#             movendo spins
# deixa os spines do topo e da direita invisíveis
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# move os spines do bottom e da esquerda
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# remove os ticks no topo e à esquerda
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
```

# Alterando característica dos tick labels

```python
t = 2*np.pi/3

fig = pl.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.plot(X, C, linewidth=2.5, label="cosseno")
ax.plot(X, S, linewidth=2.5, label="seno")

# adiciona linhas conectando coordenada x ao ponto da curva do cosseno
ax.plot(np.ones(10)*t, np.linspace(cos(t), 0, 10), 'b--')
# ponto
ax.scatter([t,], [np.cos(t),], 50)
# anota
# xy : posição do ponto anotado
# xytext : posição da anotação
# arrowpropos : propriedades do arrow da anotação ao ponto
# ver: http://matplotlib.org/1.2.1/users/annotations_guide.html
ax.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$', 
            xy=(t, np.cos(t)),
            xycoords='data',
            xytext=(t+.3, np.cos(t)-.2),
            fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
# adiciona linhas conectando coordenadas x ao ponto da curva do seno
ax.plot(np.ones(10)*t, np.linspace(0, sin(t), 10), 'g--')
ax.scatter([t,], [np.sin(t),], 50)
# anota
ax.annotate(r'$\sin(\frac{2\pi}{3}=\frac{\sqrt{3}}{2}$',
            xy=(t, np.sin(t)),
            xytext=(t+.2, np.sin(t)+.2), 
            fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2"))

#             movendo spins
# deixa os spines do topo e da direita invisíveis
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# move os spines do bottom e da esquerda
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# remove os ticks no topo e à esquerda
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# legenda
ax.legend(loc='upper left')

# alterando característica dos tick labels

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(13)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65))
```

# Types of plots

## Regular plot

```python
x = np.linspace(-np.pi, np.pi, 100, endpoint=True)
y = np.sin(2*x)

fig = pl.figure()
ax = fig.add_subplot(111)

ax.plot(x, y, color='blue', alpha=1.)

# preenche a área entre a curva e o eixo x
ax.fill_between(x, y, alpha=.35)

ax.spines['bottom'].set_position(('data', 0))
ax.spines['top'].set_position(('data', 0))
```

## Scatter plots

```python
n = 500

# n pontos em uma distribuição normal de 0 a 1
x_normal = np.random.normal(0, 1, n)
y_normal = np.random.normal(0, 1, n)
# arctan de y x, para ser utilizado como cor
# nota: engraçado, estava alternando entre o arctan2 e o arctan
#       e dando scatters diferentes! depois que percebi que o 
#       arctan recebe apenas um parâmetro. Como estava mantendo os
#       dois parâmetros, o segundo ficava como out
c = np.arctan2(y_normal, x_normal)

fig = pl.figure()

ax_normal = fig.add_subplot(1, 1, 1)
ax_normal.scatter(x_normal, y_normal, c=c)
```

## Bar plots

```python
n = 12
X = np.arange(n)
y1 = (1-X/float(n))*np.random.uniform(.5, 1., n)
y2 = (1-X/float(n))*np.random.uniform(.5, 1., n)

fig = pl.figure()
ax = fig.add_subplot(111)

# plot das barras
ax.bar(X, +y1, facecolor='#9999ff', edgecolor='white')
ax.bar(X, -y2, facecolor='#ff9999', edgecolor='white')

# labels das barras - notar o shift
for x, y in zip(X, y1):
    ax.text(x+.4, y+.05, '%.2f' % y, ha='center', va='bottom')

for x, y in zip(X, y2):
    ax.text(x+.4, -y-.15, '%.2f' % y, ha='center', va='bottom')
    
ax.set_ylim(-1.25, 1.25)
ax.set_xlim(-1, 12)

ax.set_xticks([])
ax.set_yticks([])
```

## Contour plot

```python
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
# gera um grid (n, m), sendo x de dimensão (n) e y de dimensão (m)
# repetindo x nas linhas, retornando em X
# repetindo y nas colunas, retornando em Y
# pegando X[i][j] e Y[i][j] tem-se o ponto (i, j) da grid
X, Y = np.meshgrid(x, y)

# função para retornar a 'altura' da superfície
def f(x, y):
    return (1-x/2 + x**5 + y**3)*np.exp(-x**2 - y**2)

fig = pl.figure()
ax = fig.add_subplot(111)

# contorno preenchido(filled) - com 8 divisões igualmente espaçadas
ax.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap='hot')
# contorno normal - com 8 divisões igualmente espaçadas
C = ax.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)
ax.clabel(C, inline=1, fontsize=10)

ax.set_xticks([])
ax.set_yticks([])
```

## Imshow

```python
n = 10
x = np.linspace(-3, 3, 3.5*n)
y = np.linspace(-3, 3, 3.*n)
# gera um grid (n, m), sendo x de dimensão (n) e y de dimensão (m)
# repetindo x nas linhas, retornando em X
# repetindo y nas colunas, retornando em Y
# pegando X[i][j] e Y[i][j] tem-se o ponto (i, j) da grid
X, Y = np.meshgrid(x, y)

# função para retornar a 'altura' da superfície
def f(x, y):
    return (1-x/2 + x**5 + y**3)*np.exp(-x**2 - y**2)

fig = pl.figure()
ax = fig.add_subplot(111)

# origin informa onde vai ficar o ponto [0,0]
cas = ax.imshow(f(X, Y), cmap='autumn', origin='lower')

# shrink indica a porcentagem do tamanho da barra a ser utilizado
# o tamanho default é o tamanho da reta limitada pelo min e max dos valores
pl.colorbar(cas, shrink=.80, orientation='horizontal')

ax.set_xticks([])
ax.set_yticks([])
```
