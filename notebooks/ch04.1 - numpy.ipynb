##meshgrid rs

```python
def plot_mesh(zs):
    
    plt.imshow(zs, cmap='gray_r')
    plt.colorbar()
```

```python
points = np.arange(-5, 5, 0.01)
xs, ys = np.meshgrid(points, points)
zs = np.sqrt(xs**2 + ys**2)
plot_mesh(zs)
```

```python
zs = np.random.normal(0, 1, 10000).reshape((100, 100))
plot_mesh(zs)
```

```python
zs = np.random.binomial(1, 0.7, 10000).reshape((100, 100))
plot_mesh(zs)
```

```python
zs = np.random.binomial(1, 0.9, 10000).reshape((100, 100))
plot_mesh(zs)
```

```python
zs = np.random.binomial(10, 0.7, 10000).reshape((100, 100))
plot_mesh(zs)
```

```python
zs = np.random.binomial(100, 0.2, 10000).reshape((100, 100))
plot_mesh(zs)
```

```python
zs = np.random.poisson(0.3, 10000).reshape((100, 100))
plot_mesh(zs)
```
